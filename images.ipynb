# Imports

```python
import cv2
from ultralytics import YOLO
from PIL import Image
```

# Loading Model

```python
# Load the YOLOv8 model
model = YOLO('model_data/yolov8n.pt')
```

# Predcitions

```python
img1 = cv2.imread('images/bus.jpg')
img2 = cv2.imread('images/image_1.png')
```

```python
# Predict the image
imgs = [img1, img2]
results = model.predict(imgs)
```

## Plotting the Results

```python
results_annoted_1 = results[0].plot()
results_annoted_2 = results[1].plot()
```

```python
# Display the image
img1 = Image.fromarray(results_annoted_1[:,:,::-1])
img1
```
